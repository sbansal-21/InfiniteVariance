```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def plot_binomial_pmf(n, p):
    """
    Plots the PMF of a binomial distribution with parameters n and p.

    Parameters:
    n (int): Number of trials
    p (float): Probability of success in each trial
    """
    # Generate the range of possible outcomes
    x = np.arange(0, n + 1)
    
    # Calculate the PMF for each outcome
    pmf = binom.pmf(x, n, p)
    
    # Set font type and size
    plt.rcParams['font.family'] = 'Arial'
    plt.rcParams['font.size'] = 12
    
    # Plotting
    plt.figure(figsize=(12, 6))
    plt.bar(x, pmf, color='#8B4513')
    plt.xlabel('K= Number of Successes')
    plt.ylabel('Probability of K Successes')
    plt.title(f'Binomial Distribution (n={n}, p={p})')
    plt.grid(True)
    plt.show()

# Example usage
plot_binomial_pmf(10, 0.50)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

def plot_binomial_pmf(n, probabilities):
    """
    Plots the PMF of a binomial distribution with parameters n and a list of probabilities.

    Parameters:
    n (int): Number of trials
    probabilities (list of float): List of probabilities of success in each trial
    """
    # Set font type and size
    plt.rcParams['font.family'] = 'Arial'
    plt.rcParams['font.size'] = 12
    
    # List of brown shades
    brown_shades = ['#8B4513', '#A0522D', '#D2691E', '#CD853F', '#F4A460']
    
    # Plotting
    plt.figure(figsize=(10, 6))
    
    for i, p in enumerate(probabilities):
        # Generate the range of possible outcomes
        x = np.arange(0, n + 1)
        
        # Calculate the PMF for each outcome
        pmf = binom.pmf(x, n, p)
        
        # Plot the PMF
        plt.bar(x, pmf, alpha=0.5, label=f'Probability of Success = {p}', color=brown_shades[i*3 % len(brown_shades)])
    
    plt.xlabel('K= Number of Successes')
    plt.ylabel('Probability of K Successes')
    plt.legend()
    plt.grid(True)
    plt.show()

# Example usage
plot_binomial_pmf(50, [0.50, 0.90])
```
